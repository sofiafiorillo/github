# Data Description

In order to examine the Neighborhoods of Toronto, they will be segmented to explore the common characteristics within one. Therefore, we need to know about the names and coordinates (latitude and longitude) of each of the Neighborhoods in the city. Thus, we need the population as well, in order to determine whether it is a residential area or not, which is a key factor to establish the ideal location of the grocery store. This information will be scrapped through a Wikipedia Page that contains all the information about Neighborhoods in Canada. It will therefore be filtered so taht it only contains those of Toronto. 

Thus, we also need to obtains venues of interest that near the Neighborhoods so that we can understand how the clusters are characterized. This will be obtained by leveraging information from FourSquare.com


# Methodology

Firs of all, the Data Scrapping from Wikipedia was carried out to obtain the Pandas Frame of all the Neighborhoods in Canada, so it was filtered so that it only contained those located in Toronto. A Geocoder was used to get coordinates of the city and the same was done for each of the Neighborhoods. This was crucial since this information was used in FourSquare to obtain the nearby Venues . 

The K-Means Clustering technique was used to segment the Neighborhoods into groups and then understand the shared characteristics within these segments. This was important so that we can recommend the ideal location where to establish the grocery retail. Thus, if the retail is a success we can examine Neighborhoods with similar features, which can be a good option for a second location. 



After carrying out this methodology, we chose the first cluster of Neighborhoods as an initial filter, since this group had a wide variety of Venues such as Park, Coffee Shops and Playgrounds, indicating that it could be a residential area. 




# Results

Out of the selection, the next step was to compare the Nrighborhoods of the selected cluster with the top most populated in the city. Daivisville is therefore chosen as the ideal place to locate the Grocery Shop, since it contains the necessary characteristics (it is a residential area with a wide variety of venues) and it is also very populated.
